```python
import pandas as pd
from PIL import ImageFile
from tqdm import tqdm
from keras.preprocessing import image as keras_image #modelo para preprocesamiento con keras
import numpy as np

import cv2 #openCV
import matplotlib.pylab as plt
import h5py #modulo para guardar el conjunto de datos

from sklearn.model_selection import train_test_split
```

# CLASIFICACIÓN DE IMAGENES

```python

# funcion para procesar las imagenes
def imagen_tensorflow(img_path):
    img = keras_image.load_img("data/flower_images/flower_images/" + img_path, target_size=(128, 128))
    x = keras_image.img_to_array(img)
    #print("x_ ", x.shape)
    return np.expand_dims(x, axis=0)

# funcion que crea en conjunto de datos para el modelo
def data_model(img_paths):
    list_of_tensors = [imagen_tensorflow(img_path) for img_path in tqdm(img_paths)]
    return np.vstack(list_of_tensors)
```
